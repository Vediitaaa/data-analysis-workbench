```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
sns.get_dataset_names()
```

```python
len(sns.get_dataset_names())
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df1= df[['class','alive']]
```

```python
df1
```

```python
class_en = LabelEncoder()
alive_en = LabelEncoder()
```

```python
num_class = class_en.fit_transform(df1['class'])
num_alive = alive_en.fit_transform(df1['alive'])
```

```python
df1['num_class'] = num_class
df1['num_alive'] = num_alive
```

```python
df1
```

```python
class_train, class_test, alive_train, alive_test = train_test_split(df1[['num_class']], df1[['num_alive']], train_size = 0.3)
```

```python
class_train.shape, class_test.shape, alive_train.shape, alive_test.shape
```

```python
dt=tree.DecisionTreeClassifier()
```

```python
dt
```

```python
dt.fit(class_train, alive_train)
```

```python
alive_predict = dt.predict(class_test)
```

```python
alive_predict
```

```python
ac=accuracy_score(alive_predict, alive_test)
```

```python
ac
```

```python
print(f' The Accuracy of our Decision Tree Model is {ac*100} %')
```
